```python
import pandas as pd
```

```python
# Reading the cleaned dataset

df = pd.read_csv("../data/RoadTrafficAccidentLocations_complete_clean.csv")
```

```python
df
```

```python
# Adjusting column names

df = df.rename(columns={"Unnamed: 0" : "AccidentID"})
```

```python
# Remove all duplicates based on the SeverityCategory Code

df_severity = df.drop_duplicates(subset=['AccidentSeverityCategory'], keep='last')
```

```python
# Reset the index for better formallity

df_severity = df_severity.reset_index(drop=True)
```

```python
# Remove columns that are unnecessary for table SeverityID

df_severity = df_severity.drop(columns=['AccidentID', 'AccidentType', 'AccidentType_en',
       'AccidentInvolvingPedestrian', 'AccidentInvolvingBicycle',
       'AccidentInvolvingMotorcycle', 'RoadType', 'RoadType_en', 'CantonCode',
       'MunicipalityCode', 'AccidentYear', 'AccidentMonth', 'AccidentWeekDay',
       'Weekday', 'AccidentHour', 'AccidentID'])
```

```python
df_severity
```

```python
# Save Dataframe for import into the Oracle table SeverityID

df_severity.to_csv("../data/Table_for_severityID.csv")
```

```python
# Remove all duplicates based on the AccidentType Code

df_accidentType = df.drop_duplicates(subset=['AccidentType'], keep='last')
```

```python
# Reset the index for better formallity

df_accidentType = df_accidentType.reset_index(drop=True)
```

```python
# Remove columns that are unnecessary for table AccidentID

df_accidentType = df_accidentType.drop(columns=['AccidentID',
       'AccidentSeverityCategory', 'AccidentSeverityCategory_en',
       'AccidentInvolvingPedestrian', 'AccidentInvolvingBicycle',
       'AccidentInvolvingMotorcycle', 'RoadType', 'RoadType_en', 'CantonCode',
       'MunicipalityCode', 'AccidentYear', 'AccidentMonth', 'AccidentWeekDay',
       'AccidentWeekDay_en', 'AccidentHour', 'AccidentID'])
```

```python
df_accidentType
```

```python
# Save Dataframe for import into the Oracle table AccidentID

df_accidentType.to_csv("../data/Table_for_accidentID.csv")
```

```python
# Remove all duplicates based on the RoadType Code

df_roadType = df.drop_duplicates(subset=['RoadType'], keep='last')
```

```python
# Reset the index for better formallity

df_roadType = df_roadType.reset_index(drop=True)
```

```python
# Remove columns that are unnecessary for table RoadID

df_roadType = df_roadType.drop(columns=['AccidentID',  'AccidentType', 'AccidentType_en',
       'AccidentSeverityCategory', 'AccidentSeverityCategory_en',
       'AccidentInvolvingPedestrian', 'AccidentInvolvingBicycle',
       'AccidentInvolvingMotorcycle', 'CantonCode',
       'MunicipalityCode', 'AccidentYear', 'AccidentMonth', 'AccidentWeekDay',
       'AccidentWeekDay_en', 'AccidentHour', 'AccidentID'])
```

```python
# Remove all duplicates based on the MunicipalityCode 

df_municipality = df.drop_duplicates(subset=['MunicipalityCode'], keep='last')
```

```python
# Reset the index for better formallity

df_municipality = df_municipality.reset_index(drop=True)
```

```python
# Remove columns that are unnecessary for table Municipality - Canton relation

df_municipality = df_municipality.drop(columns=['AccidentID',  'AccidentType', 'AccidentType_en',
       'AccidentSeverityCategory', 'AccidentSeverityCategory_en',
       'AccidentInvolvingPedestrian', 'AccidentInvolvingBicycle',
       'AccidentInvolvingMotorcycle', 'RoadType', 'RoadType_en',
        'AccidentYear', 'AccidentMonth', 'AccidentWeekDay',
       'AccidentWeekDay_en', 'AccidentHour', 'AccidentID'])
```

```python
# Save Dataframe for import into the Oracle table MunicipalityCanton

df_municipality.to_csv("../data/Table_for_municipality_canton.csv")
```
